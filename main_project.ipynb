```python
import os
from six.moves import urllib
import pandas as pd
import numpy as np

DOWNLOAD_ROOT = "https://www.kaggle.com/c/march-machine-learning-mania-2017/download/"
LOCAL_PATH = "datasets/"

REG_SEASON_COMPACT_URL = "RegularSeasonCompactResults.csv"
REG_SEASON_DETAILED_URL = "RegularSeasonDetailedResults.csv"
SAMPLE_SUBMISSION_URL = "SampleSubmission.csv"
SEASONS_URL = "Seasons.csv"
TEAMS_URL = "Teams.csv"
TOURNEY_COMPACT_URL = "TourneyCompactResults.csv"
TOURNEY_DETAILED_URL = "TourneyDetailedResults.csv"
TOURNEY_SEEDS_URL = "TourneySeeds.csv"
TOURNEY_SLOTS_URL = "TourneySlots.csv"

DATASET_URLS = [REG_SEASON_COMPACT_URL, REG_SEASON_DETAILED_URL, SAMPLE_SUBMISSION_URL,
                       SEASONS_URL, TEAMS_URL, TOURNEY_COMPACT_URL, TOURNEY_DETAILED_URL, 
                       TOURNEY_SEEDS_URL, TOURNEY_SLOTS_URL]


def fetch_data(dataset_urls=DATASET_URLS, 
                   download_root=DOWNLOAD_ROOT, local_path=LOCAL_PATH):
    '''Fetch datasets onto local computer'''
    if not os.path.isdir(local_path):
        os.makedirs(local_path)
    for dataset_url in dataset_urls:
        csv_path = os.path.join(local_path, dataset_url)
        urllib.request.urlretrieve(download_root+dataset_url, csv_path)
    
def load_data(dataset_url, local_path=LOCAL_PATH):
    '''Load datasets into script'''
    csv_path = os.path.join(local_path, dataset_url)
    return pd.read_csv(csv_path)

reg_season_compact = load_data(REG_SEASON_COMPACT_URL)
reg_season_detailed = load_data(REG_SEASON_DETAILED_URL)
sample_submission = load_data(SAMPLE_SUBMISSION_URL)
seasons = load_data(SEASONS_URL)
teams = load_data(TEAMS_URL)
tourney_compact = load_data(TOURNEY_COMPACT_URL)
tourney_detailed = load_data(TOURNEY_DETAILED_URL)
tourney_seeds = load_data(TOURNEY_SEEDS_URL)
tourney_slots = load_data(TOURNEY_SLOTS_URL)
```
